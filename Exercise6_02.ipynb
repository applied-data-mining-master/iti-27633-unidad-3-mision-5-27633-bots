In this exercise, we will be implementing the following activation functions using the `numpy` package: `sigmoid`, `tanh`, `relu`, and `softmax`.

1.- Import the `numpy` package as `np`

````python
import numpy as np
````

2.- Create a `sigmoid` function, that implements the sigmoid formula.

````python
def sigmoid(x): 
    return 1 / (1 + np.exp(-x))
````

3.- Calculate the result of sigmoid function on the value -1.

Output: `0.2689414213699951`

````python
sigmoid(-1)
````

7.- Create a `tanh()` function that implements the `Tanh` formula. Then plot `x` using this function.

Output:

![Figure 6.9](img/fig6_09.jpg)

````python
def tanh(x): 
    return 2 / (1 + np.exp(-2*x)) - 1
````

The shape of the `tanh` function is very similar to `sigmoid` but its slope is steeper for values close to 0. Remember, its range is between -1 and 1.

8.- Create a `relu` function that implements the ReLU formula. Then plot using `x`.

Output:

![Figure 6.10](img/fig6_10.jpg)

````python
def relu(x):
    return np.maximum(0, x)
````

````python
plt.plot(x, relu(x))
plt.show()
````

The ReLU function equals 0 when values are negative, and equals the identity function, `f(x)=x`, for positive values.

9.- Create a `softmax` function that implements the softmax formula.

````python
def softmax(list): 
    return np.exp(list) / np.sum(np.exp(list))
````

10.- Calculate the output of softmax on the list of values, $[0, 1, 168, 8, 2]$.

Output:

```
array([1.09276566e-73, 2.97044505e-73, 1.00000000e+00, 
       3.25748853e-70, 8.07450679e-73])
```

````python
result = softmax( [0, 1, 168, 8, 2]) 
result
````

As expected, the item at the third position has the highest softmax probabilities as its original value was the highest.
